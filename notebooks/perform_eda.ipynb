```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for better aesthetics
sns.set_style("whitegrid")

# Load the dataset
try:
    df = pd.read_csv('MCA_Student_Performance_500.csv')

    # --- 1. Data Loading and Basic Inspection ---
    print("--- 1. Basic Data Inspection ---")
    print("First 5 rows of the dataset:")
    print(df.head())
    print("\n" + "="*50 + "\n")

    print("Dataset Information:")
    df.info()
    print("\n" + "="*50 + "\n")

    print("Descriptive Statistics:")
    # This generates descriptive statistics that summarize the central tendency, dispersion,
    # and shape of a datasets distribution, excluding NaN values.
    print(df.describe())
    print("\n" + "="*50 + "\n")


    # --- 2. Calculate Total Marks for EDA ---
    # We will perform most of the analysis on the total marks (out of 100)
    subjects = [
    "Algorithm_&_Program_Design", "Database_Management_Systems", "Programming_with_Java", "Digital_Logic_&_Computer_Architecture",
    "Software_Engineering", "Theoretical_Computer_Science", "Advanced_Data_Structures", "Data_Communication_&_Networks",
    "Operating_Systems_&_Shell_Programming", "AI_&_ML", "Software_Testing_&_QA", "Analysis_&_Design_of_Algorithms",
    "Data_Mining_&_Warehousing", "Information_&_Cybersecurity", "Cloud_Computing"
]

    df_totals = pd.DataFrame()
    df_totals['student_id'] = df['student_id']

    for sub in subjects:
        df_totals[sub] = df[f'{sub}_Minor1'] + df[f'{sub}_Minor2'] + df[f'{sub}_EndSem']


    # --- 3. Visualize Data Distributions ---
    print("--- 2. Generating Score Distribution Plots ---")
    plt.figure(figsize=(15, 5))
    
    # Select a few key subjects to visualize their score distribution
    subjects_to_plot = [
        "Algorithm_&_Program_Design",
        "AI_&_ML",
        "Database_Management_Systems"
    ]

    for i, subject in enumerate(subjects_to_plot, 1):
        plt.subplot(1, 3, i)
        sns.histplot(df_totals[subject], kde=True, bins=20)
        plt.title(f'Distribution of Total Scores\n in {subject}', fontsize=10)
        plt.xlabel('Total Score (out of 100)')
        plt.ylabel('Number of Students')

    plt.tight_layout()
    plt.savefig('score_distributions.png')
    print("Saved 'score_distributions.png'")


    '''# --- 4. Analyze Correlations with a Heatmap ---
    print("\n--- 3. Generating Correlation Heatmap ---")
    plt.figure(figsize=(12, 10))
    
    # Calculate the correlation matrix on total scores to find relationships between subjects
    corr_matrix = df_totals[subjects].corr()
    
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=.5)
    plt.title('Correlation Heatmap of Subject Scores')
    plt.xticks(rotation=45, ha='right', fontsize=8)
    plt.yticks(rotation=0, fontsize=8)
    plt.tight_layout()
    plt.savefig('correlation_heatmap.png')
    print("Saved 'correlation_heatmap.png'")'''


    print("\n--- 4. Generating Box Plots for Outlier Detection ---")
    plt.figure(figsize=(15, 7))
    
    # Use a subset of subjects for a cleaner plot to check for outliers
    subset_subjects = [
        "Algorithm_&_Program_Design",
        "Advanced_Data_Structures",
        "AI_&_ML",
        "Data_Mining_&_Warehousing",
        "Cloud_Computing"
    ]
    
    sns.boxplot(data=df_totals[subset_subjects])
    plt.title('Score Distribution and Outliers by Subject')
    plt.ylabel('Total Score (out of 100)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('score_boxplots.png')
    print("Saved 'score_boxplots.png'")
    print("\nEDA complete. All plots have been saved as PNG files.")

except FileNotFoundError:
    print("Error: 'MCA_Student_Performance_500.csv' not found. Please make sure the file is in the correct directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```
